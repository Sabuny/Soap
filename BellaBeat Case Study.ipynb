**3.2 Data transformation**

Let's start by converting the *ActivityDate* in **daily_activity**, *ActivityHour* in **hourly_steps**, *ActivityHour* in **hourly_calories**, and *SleepDay* in **sleepday** to datetime data type.

```python
#changing data types
sleep_day["SleepDay"]=pd.to_datetime(sleep_day["SleepDay"])
hourly_steps["ActivityHour"] = pd.to_datetime(hourly_steps["ActivityHour"], format="%m/%d/%Y %I:%M:%S %p")
daily_activity["ActivityDate"]=pd.to_datetime(daily_activity["ActivityDate"])
hourly_calories["ActivityHour"]=pd.to_datetime(hourly_calories["ActivityHour"],format="%m/%d/%Y %I:%M:%S %p");
```

Remember that both **hourly_steps** and **hourly_calories** exhibit comparable structures, each with shape of (22099, 3) i.e 3 columns and 22099 rows, so we are going to merge them. The both share the Id and ActivityHour columns so we'll perform an inner join.

```python
#merge hourly_steps and hourly_calories
hourly_merged=hourly_calories.merge(hourly_steps, on=["Id","ActivityHour"],how="inner")
hourly_merged.head(5)
```

For us to know if there are in relationship with the days and the activities, we need to create new columns of days and hours.
Let's create a new column **(Hours)** and **(ActivityDate)** in the hourly_merged dataset and new column **ActivityDate** in the **daily_activity** dataset. After creating the new columns I had to change to their correct data types

```python
#getting the hours
hourly_merged["Hour"]=hourly_merged["ActivityHour"].dt.hour
#getting the dates
hourly_merged["ActivityDate"]=hourly_merged["ActivityHour"].dt.date
# change to datetime data type
hourly_merged["ActivityDate"]=pd.to_datetime(hourly_merged["ActivityDate"])
```

Moreover lets create a new column for our days from the **sleep_day, hourly_merged** and **daily_activity** datasets. We'll then drop the ActivityHour column since we don't need it.

```python
# get the days of the week in a new column
hourly_merged["Day"]=hourly_merged["ActivityDate"].dt.day_name()
# get the days of the week in a new column
sleep_day["Day"]=sleep_day["SleepDay"].dt.day_name()
# get the days of the week in a new column
daily_activity["Day"]=daily_activity["ActivityDate"].dt.day_name()
#drop the activityhour column since we don't need it
hourly_merged.drop(columns="ActivityHour",axis=1,inplace=True)
```

You may observe that our recently merged dataframe and the **sleep_day** dataframe share identical columns. Consequently, we'll combine them. Prior to this, let's modify the column name **SleepDay** in the sleep_day dataframe to align it with the **ActivityDay** column in the **hourly_merged** dataframe.

```python
#re naming the column
sleep_day = sleep_day.rename(columns={"SleepDay": "ActivityDate"})
#merging the dataframes
df_merged =hourly_merged.merge(sleep_day, on=["Id","ActivityDate","Day"], how="inner")
```

Then we'll make sure the newly merged dataframe is clean

```python
#check if the new dataframe is clean
display(df_merged.head())
print(f"Missing values present: {df_merged.isna().any()}","\n") 
print(f"Duplicates present: {df_merged.duplicated().any()}","\n") 
print(f"Number of users: {df_merged["Id"].nunique()}","\n")
```

We'll now proceed to make adjustments to the **daily_activity** dataframe. Initially, we'll introduce two new columns, namely **TotalActiveMinutes** and **TotalMinutes**, alongside **TotalActiveHours.** Following this, we'll proceed to eliminate unnecessary columns.

```python
#create new columns TotalActiveMinutes, TotalMinutes and TotalActive hours
daily_activity["TotalActiveMinutes"]=daily_activity["VeryActiveMinutes"] + daily_activity["FairlyActiveMinutes"] + daily_activity["LightlyActiveMinutes"]
daily_activity["TotalMinutes"] = daily_activity["TotalActiveMinutes"] + daily_activity["SedentaryMinutes"]
daily_activity["TotalActiveHours"] = round(daily_activity["TotalActiveMinutes"] / 60)
daily_activity["TotalDistance"]=daily_activity["VeryActiveDistance"] + daily_activity["LightActiveDistance"] + daily_activity["ModeratelyActiveDistance"]
display(daily_activity.head())
```

**4.ANALYZE**

With our data processed, it's now opportune to delve into analysis for insights. We've streamlined our datasets to three distinct dataframes: **daily_activity, merge_df** and **merge_df2.** Consequently, we'll derive insights from these dataframes. To commence, let's obtain statistical information about each of them.

```python
summary=daily_activity[["TotalSteps","TotalDistance","Calories","TotalActiveMinutes","TotalMinutes","TotalActiveHours","SedentaryMinutes"]].describe()
summary1=df_merged[["Calories","StepTotal","Hour","TotalMinutesAsleep","TotalTimeInBed"]].describe()
summary2=hourly_merged[["Calories","StepTotal","Hour"]].describe()
display(summary)
display(summary1)
display(summary2)
```

Lets look at the correlation of **hourly_merged** and **daily_activity**

```python
#Calculate correlations
corr = hourly_merged[["Calories", "Hour"]].corr()
corr1 = daily_activity[["TotalSteps", "TotalDistance", "Calories", "SedentaryMinutes"]].corr()
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
sns.heatmap(corr, ax=ax[0], annot=True, cmap="seismic")
ax[0].set_title("Correlation Heatmap: Hourly Data")
sns.heatmap(corr1, ax=ax[1], annot=True, cmap="seismic")
ax[1].set_title("Correlation Heatmap: Daily Activity Data")
plt.xticks(rotation=45)
plt.yticks(rotation=65)
plt.show()
```

Here are some noteworthy discoveries:

* There's a discernible positive correlation between steps taken and calories burned.
* Out of a total of 1218 minutes, 991 minutes were spent in a sedentary state, while approximately 228 minutes were active.
* On average, users took 7638 steps daily, and the daily calorie burn averaged 2304.
* However there's a negative correlation between Sedentary Minutes and Calories which makes sense cause you can't burn calories while at rest.

Visualizing data provides a clear and detailed view of our information. Let's create visualizations to support and strengthen our findings from the analysis phase.

  Lets visualize Total steps with Calories burnt

```python
plt.figure(figsize=(9,6))
sns.scatterplot(x="Calories",y="TotalSteps",data=daily_activity)
plt.xlabel("Calories burnt")
plt.ylabel("Total Steps covered")
plt.title("A Scatter plot of Calories burnt vs TotalSteps")
plt.show();
```

Lets visualize Calories burnt while at rest

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x="Calories",y="SedentaryMinutes",data=daily_activity)
plt.xlabel("Calories burnt")
plt.ylabel("Sedentary Minutes")
plt.title("Calories burnt by at Rest")
plt.show()
```

* These two scatterplots offer a clearer depiction of the relationship between TotalSteps and SedentaryMinutes with respect to Calories. As the number of steps taken increases, so does the calorie expenditure. Conversely, higher levels of sedentary minutes correlate with lower calorie burn.

Lets visualize the active and sedentary minutes

```python
mean_minutes = daily_activity[["SedentaryMinutes", "TotalActiveMinutes"]].mean()
mean_minutes.plot(kind="bar", figsize=(8, 6), rot=0)
plt.ylabel("Count")
plt.title("Sedentary Minutes Vs Total Active Minutes")
plt.show()
```

* There's a substantial disparity between SedentaryMinutes and TotalActiveMinutes. It's apparent that users spend significantly more time in sedentary positions compared to being actively engaged.

Lets vizualise distribution of distances by activity level

```python
summary_df = daily_activity[["VeryActiveDistance", "ModeratelyActiveDistance", "LightActiveDistance", "SedentaryActiveDistance"]].sum()
colors = sns.color_palette('pastel')
plt.figure(figsize=(8, 6))
plt.pie(
    summary_df,
    labels=summary_df.index,
    autopct="%1.0f%%",
    startangle=0,
    colors=colors, 
    shadow=False,
    explode=(0.1, 0.05, 0.05, 0.05),
    wedgeprops={'edgecolor': 'black'} 
)
plt.title("Proportion of Intensity of Exercise Activity")
plt.axis("equal")
plt.show()
```

* The chart depicts that users exercise shorter distances when they are active compared to when they are less exercise distance when active. Furthermore, during periods of rest or relaxation, they cover no distance at all.

Lets visualize and see the most and least active hours throughout the day

```python
plt.figure(figsize=(8,6))
hourly_merged.groupby(hourly_merged["Hour"])["StepTotal"].mean().plot(kind="bar", rot=0)
plt.xlabel("Time During the Day")
plt.ylabel("Total Steps")
plt.title("Active & Least Active Days During the Day")
plt.show()
```

Indeed the users are we observe that their day becomes increasingly hectic starting from 8 a.m. until 7 p.m. The least active hours span from 8 p.m. to 10 p.m. These hours likely present optimal opportunities for targeted ad campaigns to reach them effectively.

Lets look at calories burnt by the number of days in a weeek

```python
plt.figure(figsize=(8,6))
plt.bar(hourly_merged.Day, hourly_merged.Calories)
plt.title("Calories burnt at Rest")
plt.xlabel("Days")
plt.ylabel("Calories Burned")
plt.show()
```

* The graph illustrates that the highest calorie expenditure occurred on Saturday, while the lowest occurred on Monday. It's understandable that Saturday, being a weekend, offers users ample time for exercise. However, the notably low calorie burn on Monday is somewhat unexpected, considering it's typically perceived as a busy workday for many individuals. One plausible explanation could be that users are heavily occupied at their desks on Mondays, leaving little time for exercise.

Lets look at the total steps covered in each day

```python
plt.figure(figsize=(8,6))
plt.bar(daily_activity.Day, daily_activity.TotalSteps)
plt.xlabel("Week Days")
plt.ylabel("Total Steps Taken")
plt.title("Total Steps Taken by Week Days")
plt.show()
```

* We observe that fewer steps were taken by users on Monday compared to the rest of the week, indicating prolonged periods spent seated at work desks. This insight also hints at the user profile likely consisting of working-class individuals. Conversely, the highest step count occurred on Saturday and Sunday, which aligns with the expectations.
